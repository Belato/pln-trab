```python
from models.pre_processing import PreProcessing
from models.vectorization_model import VectorizationModel
from models.similarity_calculation import SimilarityCalculation
from models.visualization import Visualization
```

```python
class Input:

    def __init__(self, text):
        self.original_text = text

    def add_original_docs(self, documents):
        self.original_documents = documents
```

```python
text = "A inteligência artificial está transformando o mund0, com n0vas aplicações; Algoritmos de aprendizado de máquina sã0 essenciais para isso; A automação robótica é um dos camp0s mais interessantes da tecnologia hoje em-dia; O! processamento de8linguagem natural (PLN) permite que computadores compre+endam a linguagem humana; Chatbots e assistentes de voz sã0 um bom exemplo do PLN em ação."
```

```python
input = Input(text)
```

```python
pre_processing = PreProcessing()
vetorization_model = VectorizationModel()
similarity_calculation = SimilarityCalculation()
visualization = Visualization()
```

```python
text = pre_processing.split_in_documents(text)
input.add_original_docs(text)
```

```python
text = pre_processing.pipeline(text)
binary_vectors = vetorization_model.binary_aproach(text)
bow_vectors = vetorization_model.bow_aproach(text)
tf_idf_vectors = vetorization_model.tf_idf_aproach(text)
word2vec_vectors = vetorization_model.word2vec_aproach_sentences(text)
fasttext_vectors = vetorization_model.fasttext_aproach_sentences(text)
```

```python
binary_similarity_jd = similarity_calculation.jaccard_distance_method(binary_vectors)
binary_similarity_cs = similarity_calculation.cosine_similarity_method(binary_vectors)

bow_similarity = similarity_calculation.cosine_similarity_method(bow_vectors)

tf_idf_similarity = similarity_calculation.cosine_similarity_method(tf_idf_vectors)

word2vec_similarity = similarity_calculation.cosine_similarity_method(word2vec_vectors)

fasttext_similarity = similarity_calculation.cosine_similarity_method(fasttext_vectors)
```

```python
word2vec_similarity
```

```python
matrix_bin_jd = visualization.plot(binary_similarity_jd, input.original_documents, vec_model='binary', sim_method='jaccard_distance')
matrix_bin_cs = visualization.plot(binary_similarity_cs, input.original_documents, vec_model='binary', sim_method='cosine_similarity')
matrix_bow_cs = visualization.plot(bow_similarity, input.original_documents, vec_model='bow', sim_method='cosine_similarity')
matrix_tfidf_cs = visualization.plot(tf_idf_similarity, input.original_documents, vec_model='tf_idf', sim_method='cosine_similarity')
matrix_w2v_cs = visualization.plot(word2vec_similarity, input.original_documents, vec_model='word2vec', sim_method='cosine_similarity')
matrix_ftxt_cs = visualization.plot(fasttext_similarity, input.original_documents, vec_model='fasttext', sim_method='cosine_similarity')
```
